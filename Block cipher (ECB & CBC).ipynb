```python
"""1. Write a menu-driven program to encrypt/decrypt byte-string data using 3-bit block cipher with the following options:
                   1. ECB-mode (with appropriate padding)
                   2. XOR with CBC-mode (with appropriate padding)"""

def pad(plaintext, block_size):
    padding = block_size - (len(plaintext) % block_size)
    return plaintext + '0' * padding

def ecb_encrypt(plaintext, key):
    ciphertext = ''
    for i in range(0, len(plaintext), 3):
        block = plaintext[i:i+3]
        encrypted_block = xor(block, key)
        ciphertext += encrypted_block
    return ciphertext

def xor(a, b):
    result = ''
    for i in range(len(a)):
        result += str(int(a[i]) ^ int(b[i]))
    return result


def ecb_decrypt(ciphertext, key):
    plaintext = ''
    for i in range(0, len(ciphertext), 3):
        block = ciphertext[i:i+3]
        decrypted_block = xor(block, key)
        plaintext += decrypted_block
    return plaintext


def cbc_encrypt(plaintext, key, iv):
    ciphertext = ''
    plaintext = pad(plaintext, 3)
    prev_block = iv
    for i in range(0, len(plaintext), 3):
        block = plaintext[i:i+3]
        xored_block = xor(block, prev_block)
        encrypted_block = xor(xored_block, key)
        ciphertext += encrypted_block
        prev_block = encrypted_block
    return ciphertext

def cbc_decrypt(ciphertext, key, iv):
    plaintext = ''
    prev_block = iv
    for i in range(0, len(ciphertext), 3):
        block = ciphertext[i:i+3]
        decrypted_block = xor(block, key)
        decrypted_block = xor(decrypted_block, prev_block)
        plaintext += decrypted_block
        prev_block = block
    return plaintext.rstrip('0')

print('3-bit block cipher menu:')

while True:
     
    print('1. Encrypt message')
    print('2. Decrypt message')
    print('3. Quit')
    choice = input('Enter your choice: ')

    if choice == '3':
        print("Exiting....")
        break

    text = input('Enter the bits: ')
    key = input('Enter the 3-bit key: ')
    iv = input('Enter the 3-bit IV: ')
    
    if len(key) < 3 or len(iv) < 3:
        print("Key and IV must be at least 3 bits long. Please re-enter.")
        continue
    
    if choice == '1':
        ecb_ciphertext = ecb_encrypt(text, key)
        cbc_ciphertext = cbc_encrypt(ecb_ciphertext, key, iv)
        print('CBC Encryption:', cbc_ciphertext,"\n")

    elif choice == '2':
        cbc_decrypted = cbc_decrypt(text, key, iv)
        print('CBC Decryption:', cbc_decrypted)
        
        ecb_decrypted = ecb_decrypt(cbc_decrypted, key)
        print('ECB Decryption:', ecb_decrypted,"\n")
```
